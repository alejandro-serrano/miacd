```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
var2 = Dog("Xolo", 3)
print(type(var2))
print(var2.name, var2.age)
```

```python
class Shihtzu(Dog):
    def __init__(self, name, age):
        super().__init__(name, age)

    def run(self):
        print("Shihtzu runs")
```

```python
var10 = Shihtzu("Marilyn", 2)
```
