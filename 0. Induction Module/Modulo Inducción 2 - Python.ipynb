### key

```python
palabras = ['auto', 'Gato', 'edificio', 'Cerillo', 'licuadora']
palabras
```

```python
res1 = sorted(palabras)
res1
```

```python
cad = "HermEneGilDo"
print(cad.lower())
```

```python
res2 = sorted(palabras, key = str.lower)
res2
```

```python
alumnos = [{'nombre': 'Manuel', 'nota': 95},
           {'nombre': 'Rodrigo', 'nota': 45},
           {'nombre': 'Andrea', 'nota': 78},
           {'nombre': 'Carlos', 'nota': 85}]
alumnos
```

```python
res3 = sorted(alumnos, key = lambda d: d['nota'])
res3
```

```python
res4 = sorted(alumnos, key = lambda d: d['nombre'])
res4
```

```python
sorted(palabras)
```

```python
alumnos2 = [{'nombre': 'alberto'},{'nombre': 'Abelardo'},{'nombre': 'Carlos'},{'nombre': 'cristina'}]
res5 = sorted(alumnos2, key = lambda d: d['nombre'].lower())
res5
```

```python
sorted(palabras, key = len)
```
