Encuentra todos los n√∫meros primos hasta un n√∫mero entero n dado.

```python
# Escribe tu soluci√≥n aqu√≠    
def primos_hasta_n(n):
    primos = []
    for num in range(2, n + 1):  # empezamos desde 2 porque 0 y 1 no son primos
        es_primo = True
        for i in range(2, int(num**0.5) + 1):  # probamos divisores hasta la ra√≠z cuadrada
            if num % i == 0:
                es_primo = False
                break
        if es_primo:
            primos.append(num)
    return primos

n = int(input("Ingresa un n√∫mero entero: "))
print("N√∫meros primos hasta", n, ":", primos_hasta_n(n))
```

Escribe un programa que calcule la ra√≠z cuadrada de un n√∫mero usando el m√©todo de Newton-Raphson (sin usar math.sqrt).

```python
# Escribe tu soluci√≥n aqu√≠
def raiz_cuadrada_newton(n, iteraciones=10):
    x = n / 2.0
    for _ in range(iteraciones):
        x = 0.5 * (x + n / x)  # f√≥rmula de Newton-Raphson
    return x

n = float(input("Ingresa un n√∫mero: "))
resultado = raiz_cuadrada_newton(n)
print("La ra√≠z cuadrada aproximada de", n, "es:", resultado)
```

Dada una lista de coordenadas (x, y), almac√©nalas en tuplas y calcula cu√°l est√° m√°s cerca del origen.

```python
# Escribe tu soluci√≥n aqu√≠

coordenadas = [(3, 4), (1, 1), (7, 2), (0, 5)]

mas_cercana = coordenadas[0]

for punto in coordenadas:
    distancia_punto = punto[0]**2 + punto[1]**2
    distancia_mas_cercana = mas_cercana[0]**2 + mas_cercana[1]**2

    if distancia_punto < distancia_mas_cercana:
        mas_cercana = punto

print("La coordenada m√°s cercana al origen es:", mas_cercana)
```

Escribe una funci√≥n que reciba una lista de tuplas (nombre, edad) y devuelva la tupla de la persona m√°s vieja.

```python
# Escribe tu soluci√≥n aqu√≠
def persona_mas_vieja(lista_personas):
    if not lista_personas:
        return None  # Si la lista est√° vac√≠a

    # Suponemos que la primera persona es la m√°s vieja
    mas_vieja = lista_personas[0]

    for persona in lista_personas[1:]:
        nombre, edad = persona
        if edad > mas_vieja[1]:
            mas_vieja = persona

    return mas_vieja

personas = [("Carolina", 35), ("Ian", 11), ("Emma", 6), ("Alejandro", 41)]

persona_mayor = personas[0]

for persona in personas[1:]:
    nombre, edad = persona
    if edad > persona_mayor[1]:
        persona_mayor = persona

print(f"La persona de mayor edad es: {persona_mayor}")
```

Implementa el algoritmo de ordenamiento por burbuja.

```python
# Escribe tu soluci√≥n aqu√≠
def bubble_sort(lista):
    n = len(lista)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

# Ejemplo de uso
numeros = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", numeros)
print("Lista ordenada:", bubble_sort(numeros))
```

Cuenta la frecuencia de cada letra en una palabra.

```python
# Escribe tu soluci√≥n aqu√≠
def frecuencia_letras(palabra):
    frecuencia = {}
    for letra in palabra:
        if letra in frecuencia:
            frecuencia[letra] += 1
        else:
            frecuencia[letra] = 1
    return frecuencia

palabra = input("Introduce una palabra: ")
print(frecuencia_letras(palabra))
```

Dado un texto, devuelve las 3 palabras m√°s frecuentes usando un diccionario.

Cuenta cu√°ntas vocales tiene una frase.

```python
# Escribe tu soluci√≥n aqu√≠
def cuenta_vocales(frase):
    frecuencia_vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
    for letra in frase:
        letra = letra.lower()
        if letra in frecuencia_vocales:
            frecuencia_vocales[letra] += 1
    return frecuencia_vocales

texto = input("Introduce una frase: ")
print(f"Vocales en la frase: \n{cuenta_vocales(texto)}")
```

Escribe una funci√≥n que determine si una cadena es un pal√≠ndromo ignorando espacios y may√∫sculas.

```python
# Escribe tu soluci√≥n aqu√≠
def es_palindromo(palabra):
    palabra = palabra.replace(" ", "").lower()
    return palabra == palabra[::-1]

print(es_palindromo("ojo"))
print(es_palindromo("Anita lava la tina"))
print(es_palindromo("prueba"))
```

Escribe una funci√≥n recursiva que calcule el n-√©simo n√∫mero de Fibonacci.

```python
# Escribe tu soluci√≥n aqu√≠
def fibonacci_recursivo(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)

print(fibonacci_recursivo(7))
```
