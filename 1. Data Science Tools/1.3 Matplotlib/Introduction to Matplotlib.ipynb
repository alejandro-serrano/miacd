# Matplotlib

```python
## Libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # Range of points between 0 and 10 with 100 points distributed equidistantly
y = np.sin(x)
# Graph the 100 points with the sine value
plt.plot(x,y)
plt.title('Basic graph example')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

### Custom Graph

```python
plt.plot(x,y, color="red", linestyle="--", linewidth=2, label="sin(x)")
plt.plot(x,np.cos(x), color="blue", linestyle=":", linewidth=2, label="cos(x)")
plt.title("Custom Graph")
plt.legend()
plt.show()
```

# Subplots

```python
x = np.linspace(0, 4*np.pi, 400)
y = np.sin(x**2)
data = np.random.randn(1000)

fig, axs = plt.subplots(2,2, figsize=(8,6))
axs[0,0].plot(x,y)
axs[0,0].set_title("Plot 1")

axs[0,1].scatter(x,np.cos(x))
axs[0,1].set_title("Plot 2")

axs[1,0].hist(data, bins=20)
axs[1,0].set_title("Plot 3")

axs[1,1].bar(["A", "B", "C"], [3,5,7])
axs[1,1].set_title("Plot 4")

plt.tight_layout() # Adjust the padding between and around subplots.

plt.show()
```

## Contour

```python
def f(x,y):
    return np.sin(x)**5 + np.cos(10 + y + x) * np.cos(x)

x = np.linspace(0, 5, 10)
y = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
# plt.contour(X,Y,Z)
cs = plt.contour(X,Y,Z, levels=10)
plt.clabel(cs,inline=1,fontsize=10,fmt="%1.1f")
plt.show()
```

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise= 10, random_state=42)

plt.scatter(X,y, color="blue", alpha=0.6, label="Datos")
plt.title("Datos sin outliers")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Introducir outliers
np.random.seed(0)
X_outliers = np.random.uniform(low=-3, high=3, size=(5,1))
Y_outliers = np.random.uniform(low=-300, high=300, size=5)

X = np.vstack([X, X_outliers])
y = np.concatenate([y, Y_outliers])

plt.scatter(X,y, color="blue", alpha=0.6, label="Datos")
plt.title("Datos con outliers")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
Q1 = np.percentile(y, 25)
Q3 = np.percentile(y, 75)
IQR = Q3 - Q1
lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

mask = (y >= lim_inf) & (y <= lim_sup)
print(mask)

X_filtrado = X[mask]
y_filtrado = y[mask]

print(f"Puntos originales: {len(y)}")
print(f"Puntos despus del mtodo IQR: {len(y_filtrado)}")
```

```python
from sklearn.linear_model import LinearRegression

modelo_original = LinearRegression()
modelo_original.fit(X,y)

modelo_filtrado = LinearRegression()
modelo_filtrado.fit(X_filtrado, y_filtrado)

plt.scatter(X,y, color="gray", alpha=0.6, label="Datos Originales")
plt.scatter(X_filtrado,y_filtrado, color="blue", alpha=0.6, label="Datos sin outliers")

x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)

plt.plot(x_line, modelo_original.predict(x_line), color="red", label = "Originales")
plt.plot(x_line, modelo_filtrado.predict(x_line), color="green", label = "Sin outliers")
plt.legend()
plt.show()
plt.show()
# sjmedina@up.edu.mx
```
