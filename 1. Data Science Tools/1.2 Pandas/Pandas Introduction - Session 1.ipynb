#### DataFrame

```python
area_dict = {'California': 423, 'Texas': 695, 'New York': 141, 'Florida':170, 'Illinois': 149}
area = pd.Series(area_dict)
area
```

```python
population_dict = {'California': 3833, 'Texas': 2644, 'New York': 1965, 'Florida':1955, 'Illinois': 1288}
population = pd.Series(population_dict)
population
```

```python
states = pd.DataFrame({'population': population, 'area': area})
states
```

```python
print(states.index, states.columns)
```

```python
states['population']
```

```python
states['population']['California']
```

```python
states.loc['California']
```

```python
states.iloc[0]
```
