#### 19. Multiply a 5x3 matrix by a 3x2 matrix (real matrix product).

```python
matrix_A = np.random.rand(5, 3)
print("Matrix A (5x3):\n", matrix_A)

matrix_B = np.random.rand(3, 2)
print("\nMatrix B (3x2):\n", matrix_B)

product_matrix_dot = np.dot(matrix_A, matrix_B)
print("\nProduct matrix (using np.dot()):\n", product_matrix_dot)

product_matrix_at = matrix_A @ matrix_B
print("\nProduct matrix (using @ operator):\n", product_matrix_at)
```

#### 24. Extract the integer part of a random array of positive numbers using 4 different methods.

```python
arr = np.random.uniform(0, 10, size=(3, 3))  # Random array of positive floats
integer_part_floor = np.floor(arr)
print("Original array:\n", arr)
print("Integer part using np.floor():\n", integer_part_floor)

integer_part_astype = arr.astype(int)
print("Original array:\n", arr)
print("Integer part using astype(int):\n", integer_part_astype)

integer_part_trunc = np.trunc(arr)
print("Original array:\n", arr)
print("Integer part using np.trunc():\n", integer_part_trunc)

_, integer_part_modf = np.modf(arr)
print("Original array:\n", arr)
print("Integer part using np.modf():\n", integer_part_modf)
```

#### 30. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates.

```python
cartesian_coords = np.random.rand(10, 2) * 20 - 10

x_coords = cartesian_coords[:, 0]
y_coords = cartesian_coords[:, 1]

# r = sqrt(x**2 + y**2)
radii = np.sqrt(x_coords**2 + y_coords**2)

# theta = arctan2(y, x)
angles = np.arctan2(y_coords, x_coords)

polar_coords = np.column_stack((radii, angles))

# Print the original and converted matrices to see the results
print("Original 10x2 Cartesian Coordinates Matrix:")
print(cartesian_coords)

print("\nConverted 10x2 Polar Coordinates Matrix (Radius, Angle in radians):")
print(polar_coords)
```
