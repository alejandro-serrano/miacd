```python
def zeros_and_ones_matrix(size):
    matrix = np.zeros((size,size))
    matrix[0, :] = 1
    matrix[-1, :] = 1
    matrix[:,0] = 1
    matrix[:,-1] = 1
    return matrix

print(zeros_and_ones_matrix(9))
```

```python
def celsius_to_fahrenheit(temperatures_list):
    return temperatures_list * (9/5) + 32
temperatures = np.array([-5, 0 , 18, 23, 31, 40])

print(f"Celsius: {temperatures}")
print(f"Fahrenheit: {celsius_to_fahrenheit(temperatures)}")
```

```python
def replace_negatives(array):
    result = array.copy()
    result[result < 0] = result[result < 0] ** 2
    return result

F = np.array([4,-3,-5,1,8,9,-7,-2,0])
print(f"{F}\n")
print(replace_negatives(F))
```

```python
def chess_board_matrix(n):
    board = np.zeros((n,n))
    board[::2, 1::2] = 1
    board[1::2] = 1
    return board

print(chess_board_matrix(8))
```
