```python
def inv(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

values = np.random.randint(1,10,5)
print(values, inv(values))

big = np.random.randint(1,100,1000000)

%timeit inv(big)
%timeit (1.0/big)
```
